Define x and y

```python
normal.Amount.describe()
```

```python
x = df.drop(['Class'],axis=1)
```

```python
y = df['Class']
```

Preprocessing the data

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
con=pd.DataFrame(ss.fit_transform(x),columns=ss.get_feature_names_out())
```

split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(con,y,test_size=0.32,random_state=108)
```

```python
x_train
```

```python
x_test
```

Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

lo=LogisticRegression()
dt=DecisionTreeClassifier()
ada=AdaBoostClassifier()
#rf=RandomForestClassifier()
sc=SVC()

model_list=[lo,dt,ada,sc]
```

```python

from sklearn.metrics import accuracy_score,f1_score

for i in model_list:
    i.fit(x_train,y_train)

    y_pred_train=i.predict(x_train)
    y_pred_test=i.predict(x_test)

#     f1_train=f1_score(y_pred_train,y_train)
#     f1_test=f1_score(y_pred_test,y_test)

    acc_train=accuracy_score(y_pred_train,y_train)
    acc_test=accuracy_score(y_pred_test,y_test)

#     print('f1_score_train= ',f1_train)
#     print('--'*20)
#     print('f1_score_test= ',f1_test)
#     print('--'*20)
    print(i)
    print('accuracy_score_train= ',acc_train)
    print('accuracy_score_test= ',acc_test)
    print('diff=', acc_train-acc_test)
    print('--'*20)
```

Prediciton on test data

```python
for i in model_list:
  pred_i = i.predict(x_test)
  predss_i = pd.Series(pred_i)
  print(f'Value counts for predictions on test data from the {i} model:')
  predss_i_labels = predss_i.map({1: 'Fraud', 0: 'Normal'})
  print(predss_i_labels.sample(4))
  print(predss_i.value_counts())
  print('--'*30)
```
